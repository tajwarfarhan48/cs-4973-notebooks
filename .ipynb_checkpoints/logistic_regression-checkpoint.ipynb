```python
import pandas as pd
```

```python
df = pd.read_csv('data/insurance_data.csv')
```

```python
import matplotlib.pyplot as plt

plt.plot(df['age'], df['bought_insurance'], 'bo')
```

```python
from sklearn.linear_model import LogisticRegression

X = df['age'].values.reshape(-1, 1)
y = df['bought_insurance'].values

print(X.shape)
print(y.shape)

logreg = LogisticRegression()

logreg.fit(X, y)
```

```python
m = logreg.coef_[0][0]
c = logreg.intercept_[0]

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def pred(x):
    z = m * x + c
    return sigmoid(z)
```

```python
import numpy as np

x = range(15, 66)
y_preds = []

for value in x:
    y_preds.append(pred(value))

plt.plot(X, y, 'bo', label="Actual")
plt.plot(x, y_preds, 'rx', label="Predicted")
plt.xlabel('Age')
plt.ylabel('Probability of taking insurance')
plt.legend()
```
